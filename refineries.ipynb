# Webscraping with BeautifulSoup

```python
url = 'https://en.wikipedia.org/wiki/List_of_oil_refineries'
```

```python
response = requests.get(url)
response
#Resposta OK.
```

```python
response.content
```

```python
html = response.content
html
```

```python
#Let's do the SOUP! We will translate the content to something readable through 
soup = BeautifulSoup(html)
soup
```

```python
#We can notice that the refineries are inside 'li' html tags, so finding them all:
raw_list = soup.find_all('li')
raw_list
```

```python
#Filtering the elements of the list extracted, we will gather all elements which start with upper case letters.
refineries_list_raw = [element.text for element in raw_list if len(re.findall('^[A-Z]', element.text))>0]
refineries_list_raw
#We end up with a raw list.
```

```python
#At the end of the list, we have about 65 elements which are not refineries.
not_refineries = refineries_list_raw[-66:-1]
```

```python
#We'll do a list comprehension to eliminate everything at the bottom which is not a refinery (0:695 to eliminate a "cookie statement" at the bottom).
refinery_general_list = [element for element in refineries_list_raw[0:695] if element not in not_refineries]
refinery_general_list
```

# Processing the webscrapped into lists

```python
#Now that we have a general list, we can separate the info into:
# - refinery name
# - country
# - company name
# - production (bbl/d)
# - BONUS: the refinery status (start-up, closed/shut down/ceased, scheduled to close/to be closed, mothballed)
```

## refinery:country dictionary

```python
#In order to link each refinery to a country, we will need to explore again the soup made.
soup
```

```python
#One can notice that the country names are inside h3 tags. Algeria for example:
soup.find_all('h3')[0].text
```

```python
#Let's extract the country (inside <h3>) and the refineries (inside <li>).
pais_refinaria = soup.find_all(['h3','li'])
pais_refinaria
```

```python
countries_raw_list = [element.text for element in pais_refinaria][0:1139]
countries_raw_list
#countries_raw_list: Essa  a lista base para trabalharmos toda a anlise.
```

```python
unnecessary_list = [string for string in countries_raw_list if len(re.findall('^[\d].*', string)) > 0]
unnecessary_list
```

```python
#We've got our full list, let's replace some things that may come into our way while analyzing the data.
list1 = [string.replace('&', 'and').replace('.','').rstrip() for string in countries_raw_list if string not in unnecessary_list]
list1
```

```python
#Removing the \xa0 metacharacter we identified:
list2 = [re.sub('\xa0','',string) for string in list1][-3000:-100]
list2
#We do have some elements that are not countries neither refineries, they start with ^.
```

```python
#Identifying the data which starts with '^' and 'List'.
list2_rejects = [string for string in list2 if re.findall('^[\^]', string) or re.findall('List',string)]
list2_rejects
```

```python
#Removing the rejects from our list:
list3 = [string for string in list2 if string not in list2_rejects]
list3
```

```python
#There are some elements which should be divided by the \n.
list4 = [string.split('\n') if len(re.findall('\n',string)) > 0 else string for string in list3]
list4

list4 = []

for string in list3:
    if len(re.findall('\n',string)) > 0:
        list_of_splits = string.split('\n')
        for element in list_of_splits:
            list4.append(element)
    else:
        list4.append(string)

list4
```

```python
#We then got them divided. However the group title of some refineries remain as a refinery itself.
#Let's remove these two ('Paraguana...' and 'Upgraders...'):
names_to_remove = [element for element in list4 if re.findall('^Upgraders', element) or re.findall('^Paraguana',element)]
names_to_remove
```

```python
#Removing them from the list, we have the refineries and the countries.
count_refin_list = [element for element in list4 if element not in names_to_remove]
count_refin_list
#Very good! We've got a list of countries and refineries finally.
```

```python
#Now let's create a dictionary of these countries and refineries.
#First, let's populate the dictionary with our keys countries
count_refin_dict = {string:[] for string in count_refin_list if len(re.findall('edit',string)) != 0}
count_refin_dict
#Now we have a dictionary with the keys and empty values.
```

```python
#Now let's populate the values to the keys.
#In each value, we should have a list with the refineries from that country.
#We will have to do a loop to run through the elements in count_refin_list (which has both the countries and the refineries).

for string in count_refin_list:
    if string in count_refin_dict.keys():
        var = string
    else:
        count_refin_dict[var].append(string)
        
count_refin_dict
#Now we have a dictionary that relate refineries with their respective country.
```

```python
#Inverting the dictionary, we have a dictionary in this style:
# {refinery : country}
count_refin_dict_inv = {v: k for k, values in count_refin_dict.items() for v in values}
count_refin_dict_inv
```

```python
len(count_refin_dict_inv)
#We have 695 pairs {refinery:country}.
```

## refinery_name list

```python
#Transforming the keys of our previous dictionary into a list, we have:
raw_refinery_list = list(count_refin_dict_inv.keys())
raw_refinery_list
```

```python
len(raw_refinery_list)
```

## company_name_list

```python
#Looking for not-numeric data inside parenthesis:
raw_company_list = [re.findall('\((.*?)\)|$',string)[0] for string in raw_refinery_list]
raw_company_list
```

```python
company_name_list = raw_company_list
```

```python
len(company_name_list)
```

## daily_production_list

```python
#We are going to use the raw_refinery_list for this altogether with a smart regex pattern.
# There are 'bbl/d' and 'bbl/day' and 'barrels per day' for the same thing.
# Also, there are refineries with no data on daily production or production at all.

raw_daily_production_list = []

for string in raw_refinery_list:
    
    daily_production = []
    daily_production2 = []
    daily_production3 = []
    
    daily_production = re.findall('(\d*,?\d+,\d+)[ ]?bb', string)
    daily_production2 = re.findall('(\d*,?\d+,\d+) barrels per day', string)
    daily_production3 = re.findall('(\d*,?\d+,\d+) bpd', string)
    
    if len(daily_production) > 0:
        raw_daily_production_list.append(daily_production[0])
    elif len(daily_production2) > 0:
        raw_daily_production_list.append(daily_production2[0])
    elif len(daily_production3) > 0:
        raw_daily_production_list.append(daily_production3[0])    
    else:
        raw_daily_production_list.append('Not available')
        
raw_daily_production_list
```

```python
#Como no observamos nenhuma irregularidade, vamos atrelar a varivel final o valor:
daily_production_list = raw_daily_production_list
```

```python
len(daily_production_list)
```

# Dataframe 1: Refineries

```python
# We have the following data:
# A dictionary count_refin_dict where we have refinery:country (695 entries)
# A list production_list (695 entries)
# A list company_list (695 entries)

# We have to separate all info in 4 lists.
# refineries_list, countries_list, production_list, company_list
# Pack them all together into a dataframe.
```

```python
#Let's unzip our dictionary 'count_refin_dict_inv' and have two lists instead.
raw_refineries_list = list(count_refin_dict_inv.keys())
len(raw_refineries_list)

#Since they have too much info on this field, let's filter it by the first parenthesis or comma:
refineries_list = [re.findall('(.+?)\(|$', string)[0].strip() for string in raw_refineries_list]
refineries_list
```

```python
#Lets remove the '[edit]', which contains 6 characaters, at the end of the data column country.
raw_countries_list = list(count_refin_dict_inv.values())
countries_list = [country[:-6] for country in raw_countries_list]
len(countries_list)
```

```python
len(daily_production_list)
```

```python
len(company_name_list)
```

```python
# Therefore all columns have the same shape.
```

```python
# Now let's build our dataframe.
# dictionary with list object in values 
info = {
    'rawdata' : raw_refinery_list,
    'refineryID' : refineries_list, 
    'company' : company_name_list, 
    'production' : daily_production_list,
    'country' : countries_list
} 
  
# creating a Dataframe object  
df_refineries = pd.DataFrame(info) 
  
df_refineries
```

## status column creation inside the dataframe

```python
#Now that we have the dataframe, we can easily create another column and insert the values there
#depending on the words inside the observations in 'raw_refineries_list'.

#Words such as: closed/shut down/ceased, scheduled to close/to be closed, mothballed, etc.

#Now lets work on the list:
status_list = []

for string in raw_refineries_list:
    if len(re.findall('clos\w+|$', string)[0]) > 0:
        status_list.append('Closed')
    elif len(re.findall('shut|$', string)[0]) > 0:
        status_list.append('Closed')
    elif len(re.findall('mothb|$', string)[0]) > 0:
        status_list.append('Closed')
    elif len(re.findall('ceas|$', string)[0]) > 0:    
        status_list.append('Closed')
    elif len(re.findall('construc|$', string)[0]) > 0:    
        status_list.append('Under Construction')
    else:
        status_list.append('Operational')

status_list
```

```python
df_refineries['status'] = status_list
```
